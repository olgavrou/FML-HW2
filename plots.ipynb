```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
cv_errors = {}
for i in range(1,6):
    with open(f"errors_d_{i}.dat") as f:
        cv_errors[i] = [float(line.rstrip()) for line in f]
```

```python
x = list(range(-10,11))
figure(figsize=(16, 14), dpi=80)
for key, value in cv_errors.items():
    plt.plot(x, cv_errors[key], label=f'd={key}')
plt.legend()
plt.show()
```

```python
# consume cv error manual and print mean and stddev
cv_errors_manual = {}
dir = "."

import statistics as stat

for i in range(1,6):
    with open(f"manual_test_err_{i}.dat") as f:
        cv_errors_manual[i] = []
        for line in f:
            a = line.rstrip()
            b = a.split(",")
            c = b[0:-1]
            d = [float(x) for x in c]
            cv_errors_manual[i].append((stat.mean(d), stat.stdev(d)))
```

```python
x = list(range(-10,11))
figure(figsize=(16, 14), dpi=80)
for key, value in cv_errors_manual.items():
    means = [x[0] for x in cv_errors_manual[key]]
    stddevs1 = [x[0] - x[1] for x in cv_errors_manual[key]]
    stddevs2 = [x[0] + x[1] for x in cv_errors_manual[key]]
    plt.plot(x, means, label=f'd={key}')
    plt.fill_between(x, stddevs1, stddevs2, alpha=0.1)
plt.legend()
plt.show()
```
